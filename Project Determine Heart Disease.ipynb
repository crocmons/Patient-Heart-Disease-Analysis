Now Let's try to compare two variables vs the target variable

More Specifically we'll see how age and trestbps combined effect the target in patients over 50 years old.

```python
# create dataframe with patients over 50 years old
sns.set_theme(style='whitegrid')
over_50_years = df[df['age'] > 50]

# we create the plt here:
fig, ax = plt.subplots(figsize=(10, 6))

# I'm using scatter plot for this data:
scatter = ax.scatter(
        over_50_years['age'],
        over_50_years['trestbps'],
        c = over_50_years['target']     
)
# Customize the plot here:
ax.set(title='Heart Disease and Patients Blood Pressure Levels',
       xlabel='age',
       ylabel='Blood Pressure'
      );
ax.legend(*scatter.legend_elements(), title='Target')

# Add a meanline - (average) here:
ax.axhline(over_50_years['trestbps'].mean(),
           linestyle=':'
          );
```

```python
# Reproduce the same figure as above with the "grayscale" style

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(over_50_years["age"], 
                     over_50_years["chol"], 
                     c=over_50_years["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax.axhline(over_50_years["chol"].mean(),
           linestyle=":");
```

change the dot color so that the data point makes more visible and clear

let's change the cmap parameter of scatter() plot as well as the color parameter of axhline() to fix it.

### Replot the same figure as above except change the cmap parameter of scatter() to 'winter'
#### Also change the color param of axhline() to 'red'

```python
# create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# plot this data:
scatter = ax.scatter(
    over_50_years['age'],
    over_50_years['chol'],
    c=over_50_years['target'],
    cmap='winter'
)

# customize the dataplot
ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel = 'Age',
       ylabel = 'Cholesterol'
      );

ax.legend(*scatter.legend_elements(), title='Target')

#Add a meanline:
ax.axhline(over_50_years['chol'].mean(),
          linestyle=':',
          color = 'orange'
          );
```
